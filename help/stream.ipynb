```python
import speech_recognition as sr
import pyttsx3 as ttx
import pywhatkit
import datetime

class VoiceAssistant:
    def __init__(self):
        self.listener = sr.Recognizer()
        self.engine = ttx.init()
        self.engine.setProperty('voice', 'french')

    def parle(self, text):
        self.engine.say(text)
        self.engine.runAndWait()

    def ecoute(self):
        try:
            with sr.Microphone() as source:
                print("Parle")
                voix = self.listener.listen(source)
                command = self.listener.recognize_google(voix, language='fr-FR')
                return command
        except:
            return ""

    def assistante(self):
        command = self.ecoute()
        print(command)
        if 'bonjour' in command:
            self.parle('Bonjour frero, tranquille ou quoi?')
        elif 'oui ça va et toi' in command:
            self.parle('Oh tu sais,  ça va de mon côté. Du coup, tu as besoin de quoi?')
        elif 'mets la chanson de' in command:
            chanteur = command.replace('mets la chanson de', '')
            print(chanteur)
            pywhatkit.playonyt(chanteur)
        elif 'il est quelle heure' in command:
            heure = datetime.datetime.now().strftime('%H:%M')
            print('Il est : ' + heure)
            self.parle(f'Il est : {heure}')
        elif "comment tu t'appelles" in command:
            self.parle('Je réponds au nom de Alix C137')

# Create an instance of the VoiceAssistant class
assistant = VoiceAssistant()

# Call the assistante method in a loop
while True:
    assistant.assistante()

```

```python
import streamlit as st
from streamlit_mic_recorder import mic_recorder, speech_to_text

# Initialiser l'état de session si nécessaire
state = st.session_state
if 'text_received' not in state:
    state.text_received = []

# Créer deux colonnes pour afficher le texte reconnu et le bouton d'enregistrement
c1, c2 = st.columns(2)
with c1:
    st.write("Convert speech to text:")
with c2:
    # Utiliser la fonction speech_to_text pour convertir la parole en texte
    text = speech_to_text(language='en', use_container_width=True, just_once=True, key='STT')

# Ajouter le texte reconnu à la liste si un texte a été reçu
if text:
    state.text_received.append(text)

# Afficher tous les textes reconnus précédemment
for text in state.text_received:
    st.text(text)

# Enregistrer l'audio à partir du microphone de l'utilisateur
audio = mic_recorder(start_prompt="⏺️", stop_prompt="⏹️", key='recorder')

# Si de l'audio a été enregistré, le jouer
if audio:
    st.audio(audio['bytes'])
```

```python
import streamlit as st
from streamlit_mic_recorder import mic_recorder, speech_to_text

# Initialiser l'état de session pour gérer les dialogues
state = st.session_state
if 'dialogue' not in state:
    state.dialogue = ""

def start_dialogue():
    """Commence un nouveau dialogue."""
    state.dialogue = "Bonjour! Comment puis-je vous aider aujourd'hui?"

# Fonction pour traiter la réponse de l'utilisateur
def handle_response(response):
    """Traite la réponse de l'utilisateur et génère une nouvelle question."""
    # Ici, vous pouvez ajouter une logique plus complexe pour répondre aux différentes réponses
    # Pour cet exemple, nous allons simplement renvoyer une réponse générique
    return f"Merci pour votre réponse {response}. Quel est votre prochain souhait?"

# Bouton pour démarrer le dialogue
start_button = st.button("Parler")
if start_button:
    # Commencer un nouveau dialogue
    start_dialogue()

    # Enregistrer l'audio et convertir en texte
    audio = mic_recorder(start_prompt="Dites quelque chose...", stop_prompt="Arrêtez.", key='recorder')
    if audio:
        text = speech_to_text(language='fr', use_container_width=True, just_once=True, key='STT')
        
        # Traiter la réponse de l'utilisateur
        new_question = handle_response(text)
        state.dialogue += f"\nVous: {text}\nAssistant: {new_question}"
        st.text_area("Dialogue", value=state.dialogue, height=200)

# Afficher le dialogue actuel
st.text_area("Dialogue", value=state.dialogue, height=200)
```

```python
import tkinter as tk
import speech_recognition as sr

def listen_and_respond():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        audio = r.listen(source)
        try:
            text = r.recognize_google(audio)
            print("You said: {}".format(text))
            # Ici, vous pouvez ajouter la logique pour répondre à la commande vocale
        except:
            print("Sorry, I did not get that")

root = tk.Tk()
frame = tk.Frame(root)
frame.pack()

button = tk.Button(frame, text="Listen", command=listen_and_respond)
button.pack(side=tk.LEFT)

root.mainloop()
```
